```python
data = pd.read_excel('PLZ_Verzeichnis_Post.xlsx')
data.head()
```

Transform the data

```python
df = data.copy()
df = df[['POSTLEITZAHL','KANTON', 'SPRACHCODE','Geo Shape', 'Geokoordinaten']]
df = df.dropna(subset=['Geokoordinaten']) 
df = df.rename({'POSTLEITZAHL': 'PLZ'}, axis=1)
df = df.rename({'KANTON': 'Kanton_Kuerzel'}, axis=1)
df = df.drop_duplicates(subset=['PLZ'])
df[['Breitengrad', 'Längengrad']] = df.Geokoordinaten.str.split(",", expand = True)
df.drop('Geokoordinaten', axis=1, inplace=True)
df.head()
```
